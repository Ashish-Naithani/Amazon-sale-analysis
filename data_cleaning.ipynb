```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_excel('Amazon_Sale_Report.xlsx')
```

```python
df1.head(2)
df2=df1
```

```python
df1=df1.drop('Unnamed: 22',axis=1)
df1=df1.drop_duplicates()
df1.head()
```

```python
df2=df1.copy()
#df2.head()
```

```python
#df1=df2.copy()
```

```python
df1['Courier Status']=df1['Courier Status'].fillna('Shipped')
df1['currency']=df1['currency'].fillna('INR')
df1['Amount']=df1['Amount'].fillna(df1['Amount'].mean())
```

```python
df1=df1.fillna(
    {
        'Courier Status':'Shipped',
        'currency':'INR',
        'Amount':df1['Amount'].mean(),
        'ship-city':'BENGALURU',
        'ship-state':'KARNATAKA',
        'ship-postal-code':201301.0,
        'ship-country':'IN',
        'promotion-ids':df1['promotion-ids'].value_counts().idxmax(),
        'fulfilled-by':'Easy Ship'
    }
)
```

```python
with pd.ExcelWriter('Amazon_Sale_Reports.xlsx') as writer:
    df1.to_excel(writer,sheet_name="Amazon Sale Report",index=False)
```
